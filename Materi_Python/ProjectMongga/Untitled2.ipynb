```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4,5]
y= [0,1,4,5,8,11]
z = [3,4,5,1,2,4]
```

```python
plt.figure(figsize=(5,5),dpi=100)
```

```python
plt.plot(x,y, label="2x", color = "red", linewidth = 3, marker ="^", linestyle = "--", markersize = 5, markeredgecolor = "blue")
```

```python
plt.close()
```

```python
plt.plot(x,y, label="2x", color = "red", linewidth = 3, marker ="^", linestyle = "--", markersize = 5, markeredgecolor = "blue")
plt.legend()
```

```python
plt.plot(x,z, "bo", label="z", linewidth = 3, markersize = 10, markeredgecolor = "blue")
```

```python
plt.plot(x,y,"ro-", label ="2x")
plt.xlabel("sb.x")
plt.xticks([2,4,6,8,10])
plt.yticks([2,4,6,8,10])
plt.ylabel("sb.y")
plt.legend()
plt.title("ini grafik pertama", fontdict = {"fontname":"Comic Sans MS","fontsize":20})
plt.savefig("mygraph.png",dpi=300)
```
